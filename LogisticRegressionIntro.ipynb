# proje amacı
bir banka data setim var verilen verilere göre muşterinin abone olup olmayacağını hesaplayna bir algoritma kurmaya çalışıyorum. abone olur veya olmaz binary bir işlem bu yuzden logistic regresyon kullanıyorum.

```python
import  numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("6-bank_customers.csv")
```

```python
df.columns
```

```python
df.head()
```

```python
df.info()
```

```python
X =df.drop("subscribed",axis=1)
y = df["subscribed"]
```

bana data setim çok temiz ve standartize bir halde geldiği için ben onu transform fit transfom haline getirmeye uğraşmadım

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=15)
```

```python
from sklearn.linear_model import LogisticRegression
logistic = LogisticRegression() # burası logistic regresyon için model oluşturma kısmı
logistic.fit(X_train,y_train)
y_pred = logistic.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix , accuracy_score, classification_report
#classification scorrun içinde f1 precion ve diğer ogrendiğimiz scorlar var
```

```python
score = accuracy_score(y_pred,y_test) # burasıda regresyon verilerimiz test etme kısmı
print("score: " , score)
print(classification_report(y_pred,y_test))
print("confusion matrix")
print(confusion_matrix(y_pred,y_test))
```

# hyperparameter tuning
benim data setim bu ders için oluşturuldugu için hem temiz hem de logistic regresyonun parametreleri ile oynamama gerek kalmayacak şekilde en iyi sonucu verdi ama benim gerçek data setlerinde en iyi sonucu bulmak için logistic regresyonun parametrelerini bilmem ve kullanmam gerekir bunun için dökümantasyon okuyup parametreleri öğrenmeliyim.

```python
model = LogisticRegression()
```

```python
penalty =["l1", "l2" , "elasticnet"] #data setimdeki verilere göre over fitting engellemek veya gereksiz sutun silme işlemi listeye denemek istediklerimi ekliyorum. hepsini deneyecek hangisi faydalıysa onu kullanacak
c_values= [100,10,1,0.1,0.01] # ceza değeri penaltylaerin lamda oranları. oncelikle hangi aralıkta oldugunu bulmak için bu geniş aralıgı veriyorum. hangi aralıkta oldugunu bulduktan sonra aralıgı degiştirebilirim
solver=['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga'] #optimizasyonun hangi algoritma tarafından yapılacağını söyler bir denlememi var onun için kullanacağı parametreleri falan algoritma türüne göre seçiyor
```

```python
params=dict(penalty=penalty,C=c_values,solver=solver)
params # bunu da dictuanery hale getirme sebebim senin yerine bu işi yapacak metodların bunu öyle istmesi
```

```python
from sklearn.model_selection import StratifiedKFold , GridSearchCV # surada kullanıdıgımız seylerin nasıl çalıştıgını tam anlamadım sadec ama ne yaptıgımızı anladım
```

```python
cv=StratifiedKFold(n_splits=5) #cross- validation için kullanacagım default k degeri 5
# bunu GridSearchCV böyle istediği için böyle yazıyoruz
```

```python
grid = GridSearchCV(estimator=model,param_grid=params,cv=cv,scoring="accuracy",n_jobs=-1) # verdiğim parametrelere göre en uygun modelin ne olacağını söyluyor amacı bu nasıl yapıyor bilmiyorum yapıyor işte
#Anladım ya modeli veriyrousn en başta senğn logistic regresyonda default oluşturdugun şeyeri
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_  #bu kod  bana verdiğim tüm o degerler içinden ennn uygun olanı verdi
```

```python
grid.best_score_
```

```python
y_pred = grid.predict(X_test)
```

```python
score = accuracy_score(y_pred,y_test)
print("score: " , score)
print(classification_report(y_pred,y_test))
print("confusion matrix")
print(confusion_matrix(y_pred,y_test))
```
