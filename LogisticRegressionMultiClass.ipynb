siber güvenlikle alakalı verilen data setine bakarak siber saldırı olup olmadığını sınflandırırız. sınıflandırma için ise Logistic Regression kullanırız.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("7-cyber_attack_data.csv")
```

```python
# kolon açıklaması gelen istekleri analiz ederek siber saldırı olup olmadıgını anlamak için kullanılır
```

```python
df.info() #tertemiz data
```

```python
df.tail()
```

```python
df.describe()
```

datam temiz hemen x ve y kolonlarıma ayırma işlemi yapıyorum

```python
X = df.drop("attack_type",axis=1 )
y = df["attack_type"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 15)
```

```python
from sklearn.linear_model import LogisticRegression
logistic = LogisticRegression()
logistic.fit(X_train, y_train)
y_pred = logistic.predict(X_test)
```

```python
y_pred # benim için multi class çalıştı yanı binary değil
```

```python
from  sklearn.metrics import confusion_matrix, accuracy_score, classification_report
# oluşturdugum modeli test etmek için y_prd ve y_testlerimi kullancagım
print("score ",accuracy_score(y_pred,y_test))
print("classification report ",classification_report(y_pred,y_test))
print("confusion matrix : \n ",confusion_matrix(y_pred,y_test))
```

```python
penalty = ["l1", "l2" , "elasticnet"]
c_values = [100,10,1,0.1,0.01]
solver = ["newton-cg", "lbfgs", "liblinear", "sag", "saga", "newton-cholesky"]
```

```python
params = dict(penalty=penalty,solver=solver ,C= c_values) #✔ GridSearch için parametre sözlüğü oluşturuyorsun

#GridSearch tüm kombinasyonları tek tek dener.
```

```python
from sklearn.model_selection import GridSearchCV,StratifiedKFold
```

```python
cv = StratifiedKFold()
```

```python
model = LogisticRegression()
```

```python
grid = GridSearchCV( estimator= model ,param_grid=params, cv=cv, scoring="accuracy" )
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_score_
```

```python
grid.best_params_
```

```python
y_pred = grid.predict(X_test) #heee bu benin logistic regresyon sonucu elde ettiğim  y
```

```python
print(" score " , accuracy_score(y_pred,y_test ))
print( "classification report ",classification_report(y_pred,y_test))
print("confusion matrix : \n ",confusion_matrix(y_pred,y_test))
```

# one vs rest
OneVsOneClassifier her sınıfı birbiir ile karılaştırılır küçük data setleri için verimlirdir buyuklerde maliyet artar            OneVsRestClassifier her modelde kendisi ve diğerleir il karşılaştıma olur en iyi sonucu veren sonuç seçilir


```python
from sklearn.multiclass import OneVsOneClassifier, OneVsRestClassifier
```

```python
onevsonemodel =OneVsOneClassifier(LogisticRegression())
onevsrestmodel = OneVsRestClassifier(LogisticRegression())
```

```python
onevsonemodel.fit(X_train,y_train)
y_pred = onevsonemodel.predict(X_test)
print(" score " , accuracy_score(y_pred,y_test ))
#modelin başarısını gösterir
print( "classification report ",classification_report(y_pred,y_test)) # her sınıf için başarısını hesaplar
#recall gerçekte modelin o sınıfa ait verilerin ne kadarını yakalayabildiğini gösterir
print("confusion matrix : \n ",confusion_matrix(y_pred,y_test))
#hatalı değerleri gösterir
```

```python
onevsrestmodel.fit(X_train,y_train)
y_pred = onevsrestmodel.predict(X_test)
print(" score " , accuracy_score(y_pred,y_test ))
print( "classification report ",classification_report(y_pred,y_test))
print("confusion matrix : \n ",confusion_matrix(y_pred,y_test))
```
