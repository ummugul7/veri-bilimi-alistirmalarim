```python
import  numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df= pd.read_csv("13-car_evaluation.csv")
```

```python
df.head()
```

```python
col_names = ["buying", "maint", "doors", "persons", "lug_boot", "safety", "class"] #class isimleri hatalı olduğu için önce datayı temizleme işlemi yaptım
```

```python
df.columns=col_names
```

```python
df.head()
```

```python
for col in df.columns:
    print(df[col].value_counts())
```

```python
df["doors"].unique() #5den fazla araçları da 5e atayıp kategorik değer yapmaktan kurtarıp numeric değer e çevirmemizi sağlamış olacak
```

```python
df['doors'] = df['doors'].replace('5more','5') #
```

```python
df["doors"] =df["doors"].astype(int)
```

```python
df["persons"].unique() #burada da more yerine 5 demeyi tercih ediyoruz yine aynı sebepten dolayı
```

```python
df["persons"] = df["persons"].replace("more" ,"5")
```

```python
df["persons"] = df["persons"].astype(int)
```

```python
X= df.drop("class" , axis=1) #encoding yapmadan önce data ayırma işlemi yapıyoruz
y = df["class"]
```

```python
from sklearn.model_selection import train_test_split
X_train ,X_test, y_train,y_test= train_test_split (X, y, test_size=0.3, random_state=15)
```

```python
# ordenal encoding  kolonlarıma göre rakam yükseldikce kategorik değerin yükselmesinin bir mantığı var.
```

```python
df.columns
```

```python
from sklearn.preprocessing import OrdinalEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

```python
categorical_cols = ["buying", "maint", "lug_boot", "safety"]
numerical_cols = ["doors", "persons"]

ordinal_encoder= OrdinalEncoder( categories=[
      ["low", "med", "high", "vhigh"], #buying
    ["low", "med", "high", "vhigh"], #maint
    ["small", "med", "big"], #lug_boot
    ["low", "med", "high"] #safety
])

# aynı anda birden fazla transaformasyonu yapar
preprocessor =ColumnTransformer( #remainder parametresi passthrough yapmazsa dfin içinde buldugu kolonları drpo edebilir  geri kalanı aynı tutması için bunu yapıyorum
    transformers=[("transformatinon_name", ordinal_encoder, categorical_cols) ],remainder="passthrough"
)
X_train_transform =preprocessor.fit_transform(X_train)
X_test_transform =preprocessor.transform(X_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

notlarda gini impurity ve entropy diye bir şey vardı ya classifair için ağacın nasıl dallancağını gösteren o 

```python
tree_model=DecisionTreeClassifier(criterion="gini" ,max_depth=3 , random_state=15) # max_depth burada ne kadar asağı ineceğini söyler
tree_model.fit(X_train_transform,y_train)
```

```python
y_pred = tree_model.predict(X_test_transform)
```

```python
from sklearn.metrics import confusion_matrix , classification_report, accuracy_score
print("Confusion Matrix \n",confusion_matrix(y_test,y_pred))
print("Classification Report",classification_report(y_test,y_pred))
print("Accuracy",accuracy_score(y_test,y_pred))
```

```python
plt.figure(figsize = (12,8))
from sklearn import tree
columns_name=categorical_cols +numerical_cols
tree.plot_tree(tree_model.fit(X_train_transform,y_train),feature_names=columns_name)
```

hyperprameting tuning

```python
#Kod, buradaki her bir seçeneği diğerleriyle kombine ederek deneyecek.


param ={
    "criterion":["gini","entropy" ,"log_loss"], #karar ağacında dallara aayırmak için kullanılan metrikler
    "splitter":["best","random"],
    "max_depth":[3,4,5,6,12,None], #ağacın max derinliği
     "max_features":["sqrt","log2 " ,None] #her bölümde kaç özelliğe bakacak
}
```

```python
from sklearn.model_selection import GridSearchCV
# estimator hangi algoritmayı deneyeceğini söyler
# cros valid için kaç parçaya ayrılacağını söyler
grid = GridSearchCV(estimator=DecisionTreeClassifier(), param_grid=param,cv=5,n_jobs=-1 ,scoring="accuracy" )
import  warnings
warnings.filterwarnings("ignore")
grid.fit(X_train_transform,y_train)
```

```python
grid.best_params_
```

```python
y_pred = grid.predict(X_test_transform)
```

```python
print("Confusion Matrix \n",confusion_matrix(y_test,y_pred))
print("Classification Report",classification_report(y_test,y_pred))
print("Accuracy",accuracy_score(y_test,y_pred))
```

```python
tree_model_new = DecisionTreeClassifier(criterion="log_loss" ,max_depth=12 , splitter="best" ,max_features=None)
```

```python
tree_model_new.fit(X_train_transform,y_train)
y_pred = tree_model_new.predict(X_test_transform)
```

```python
tree.plot_tree(tree_model_new.fit(X_train_transform, y_train),feature_names=columns_name)
```
