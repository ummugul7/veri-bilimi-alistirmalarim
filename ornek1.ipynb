```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.pyplot import xlabel
```

```python
data =pd.read_csv("13-WineQT.csv")
```

```python
data.describe()
```

```python
data.head()
```

```python
data["quality"].max()
```

```python
sns.heatmap(data.corr(),annot=True)   #bileşenlerin aralarındaki Korelasyon haritası
```

```python
sns.boxplot(data=data,x="quality",y="alcohol")
plt.xlabel('quality')
plt.ylabel('alcohol')
plt.show()
```

```python
sns.set_style("whitegrid")
sns.scatterplot(data =data ,x='fixed acidity',y='volatile acidity' , hue='quality')
plt.xlabel('fixed acidity')
plt.ylabel('volatile acidity')
plt.title('fixed acidity vs volatile acidity')
plt.show()
```

```python
plt.scatter(data["residual sugar"],data["density"])
plt.xlabel('residual sugar')
plt.ylabel('density')
plt.title('residual sugar vs density')
plt.show()
```

```python
plt.scatter(data["pH"],data["alcohol"])
plt.xlabel('pH')
plt.ylabel('alcohol')
plt.title('pH vs alcohol')
plt.show()
```

```python
plt.hist(data["pH"])
plt.show()
```

```python
data.columns
```

```python
plt.figure(figsize=(12,10))

columns = [
    'pH', 'alcohol', 'quality', 'density',
    'fixed acidity', 'volatile acidity', 'citric acid',
    'residual sugar', 'chlorides', 'free sulfur dioxide',
    'total sulfur dioxide', 'sulphates'
]

for i, col in enumerate(columns, 1):  # enumerate → (index, column)
    plt.subplot(3, 4, i)
    plt.title(col)
    plt.hist(data[col], bins=20, color='skyblue', )
    plt.tight_layout()

plt.show()
```

```python
sns.set_style("whitegrid")
sns.regplot(data=data,x="density",y="alcohol")
plt.xlabel('density')
plt.ylabel('alcohol')
plt.title('density vs alcohol')
plt.show()
```

```python
selected_cols = ['alcohol', 'quality', 'residual sugar','fixed acidity']

plt.figure(figsize=(10,6))
sns.boxplot(data=data[selected_cols])
plt.title("Aykırı Değer Analizi (Outlier Detection)")
plt.xlabel("Özellikler")
plt.ylabel("Değer Aralıkları")
plt.show()
```
